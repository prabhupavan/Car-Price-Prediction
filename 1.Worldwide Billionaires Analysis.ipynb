# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading The Data

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/Billionaire.csv")
data
```

```python
print(data.head())
```

```python
# Checking For any missing values

data.isnull().sum()
```

```python
# The dataset has 79 missing values in the AGE column, let's remove these rows.

data = data.dropna()
```

```python
# The Networt column in this dataset has a $ sign at the beginning as B at the end.
# We need to remove these signs and convert the networth column to float.as_integer_ratio
data['NetWorth'] = data['NetWorth'].str.strip('$')
data['NetWorth'] = data['NetWorth'].str.strip('B')
data['NetWorth'] = data['NetWorth'].astype(float)
```

# Visualization

```python
# Let's have a look at the top 10 billionairs according to their networth

df = data.sort_values(by=['NetWorth'], ascending=False).head(10)
plt.figure(figsize=(20,10))
sns.histplot(x='Name', hue='NetWorth', data=df)
plt.show()
```

```python
# The TOP 5 Domains with the most number of Billionairs.

domain = data['Source'].value_counts().head()
index = domain.index
sources = domain.values
colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(7,7))
plt.pie(sources, labels=index, colors=colors)

# Cutting the circle from the center.
center_circle = plt.Circle((0,0), 0.5, color='white')
fig = plt.gcf() # gcf ---> Get Current Figure
fig.gca().add_artist(center_circle)
plt.rc('font', size=12)     # rc ---> Run Time Configuration
plt.title('The TOP 5 Domains with the most Billionairs', fontsize=20)   
plt.show()
```

```python
# The TOP 5 Industries with the most number of Bilionairs.

domain = data['Industry'].value_counts().head()
index = domain.index
industries = domain.values
colors = ["skyblue", "yellowgreen", 'tomato', "blue", "red"]
plt.figure(figsize=(7,7))
plt.pie(industries, labels=index, colors=colors)

# Cutting the circle from the center.
center_circle = plt.Circle((0,0), 0.5, color='white')
fig = plt.gcf() # gcf ---> Get Current Figure
fig.gca().add_artist(center_circle)
plt.rc('font', size=12)     # rc ---> Run Time Configuration
plt.title('The TOP 5 Industries with the most Billionairs', fontsize=20)   
plt.show()
```
